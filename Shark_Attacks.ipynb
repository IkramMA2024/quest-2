```python
#SPECIES CLEANING

import re
from collections import Counter
# Renommer la colonne 'Species ' en 'Species'
df_shark.columns = ['Species' if col.startswith('Species ') else col for col in df_shark.columns]

#Convertir toutes les valeurs en miniscules
df_shark['Species']=df_shark['Species'].str.lower()

all_shark_species=''.join(df_shark['Species'].dropna())

# Utiliser une expression régulière pour extraire des mots
words = re.findall(r'\b\w+\b', all_shark_species)

# Compter la fréquence des mots
word_counts = Counter(words)

# Afficher les mots les plus communs
common_words = word_counts.most_common(20)
print("Mots les plus communs :")
print(common_words)

# Calculer le nombre de lignes contenant des mots clés spécifiques
keywords = [word for word, count in common_words]
pattern = '|'.join(keywords)
lines_with_keywords = df_shark['Species'].str.contains(pattern, na=False).sum()

# Nombre total de lignes
total_lines = df_shark.shape[0]

# Afficher le résultat
print("\nNombre total de lignes : ", total_lines)
print("Nombre de lignes contenant des mots clés : ", lines_with_keywords)
print("Nombre de lignes ne contenant pas de mots clés : ", total_lines - lines_with_keywords)
```
